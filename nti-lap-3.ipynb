```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=[50,70,100,120,150]
```

```python
y=[150,200,300,350,400]
```

```python
x=np.array(x).reshape(-1,1)
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
print("slope", model.coef_)
print("intercept", model.intercept_)
```

```python
x_new= np.array([[1.75]])
y_predict= model.predict(x_new)
print("Prediction for 1.75:", y_predict)
```

```python
y_predicted_original= model.predict(x)
```

```python
plt.scatter(x, y, color='blue', label='Original data')
plt.plot(x, y_predicted_original, color='red', label='Regression line')
plt.xlabel('house size(m^2)')
plt.ylabel('price($1000)')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
y
```

```python
y_predict
```

```python
score = r2_score(y, y_predicted_original)
mse = mean_squared_error(y, y_predicted_original)
```

```python
print("R^2 score:", score)
print("Mean Squared Error:", mse)
```
