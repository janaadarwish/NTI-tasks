```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[3,21,22,34,54,34,55,67,89,99]
```

```python
y=[1,10,14,34,44,36,22,67,79,90]
```

```python
x= np.array(x)
y= np.array(y) 
```

```python
def model(w,b,x):
    return w*x + b
```

```python
def loss_func(x,w,b,y):
    num= len(x) #number of x data ===> len() returns num of x data
    prediction= model(w,b,x)
    
    return (0.5/num)*(np.square(prediction-y)).sum() #loss equation
```

```python
# Gradient descent step
def optimize(w, b, x, y, lr=0.0001):
    num = len(x)
    prediction = model(w, b, x)
    dw = (1/num) * np.dot((prediction - y), x)
    db = (1/num) * np.sum(prediction - y)
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
def iterate(times, x, w, b, y):
    error = []
    for i in range(times):
        error.append(loss_func(x, w, b, y))
        w, b = optimize(w, b, x, y)
    return w, b, error
```

```python
w=np.random.rand()
```

```python
b=np.random.rand()
```

```python
w,b,error= iterate(6,x,w,b,y)
```

```python
prediction = model(w, b, x)
```

```python
plt.scatter(x, y, color="blue")
plt.plot(x, prediction, color="red")
plt.title("Linear Regression (Gradient Descent)")
plt.show()

print(f"Final slope (w): {w}")
print(f"Final intercept (b): {b}")
```
